```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [7,8,9,10,11,12]
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('First Graph')
plt.plot(x,y)
```

```python
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('First Graph')
plt.plot(x,y, color = 'red')
```

```python
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('First Graph')
plt.plot(x,y, color = 'red', marker = '*')
```

```python
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('First Graph')
plt.plot(x,y, color = 'red', marker = '*', markersize = 10)
```

```python
x = [1,2,3,4,5,6]
plt.plot(x)
```

```python
x = [1,2,3,4,5,6]
y = [7,8,9,10,11,12]

plt.plot(x,y, 'r') #String Argument
```

```python
x = [1,2,3,4,5,6]
y = [7,8,9,10,11,12]

plt.plot(x,y, 'r*') #String Argument
```

```python
x = [1,2,3,4,5,6]
y = [7,8,9,10,11,12]

plt.plot(x,y, 'r*', markersize = 15) #String Argument
```
