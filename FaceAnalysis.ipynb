```python
import cv2  #OpenCV
import matplotlib.pyplot as plt
import face_recognition as fc
```

```python
# inline magic
%matplotlib inline
```

```python
image1 = plt.imread('images/BarackObama1.jpg')
plt.imshow(image1)
```

```python
image2 = plt.imread('images/SteveJobs.jpg')
plt.imshow(image2)
```

```python
face_locations1 = fc.face_locations(image1)[0]
print(face_locations1)
```

```python
image1_x1y1 = (face_locations1[3], face_locations1[0])
image1_x2y2 = (face_locations1[1], face_locations1[2])

print(image1_x1y1)
print(image1_x2y2)
```

```python
cv2.rectangle(image1, image1_x1y1, image1_x2y2, (255, 0, 0), 8)
plt.imshow(image1)
```

```python
face_locations2 = fc.face_locations(image2)[0]
print(face_locations2)
```

```python
image2_x1y1 = (face_locations2[3], face_locations2[0])  #x1y1
image2_x2y2 = (face_locations2[1], face_locations2[2])  #x2

print(image2_x1y1)
print(image2_x2y2)
```

```python
cv2.rectangle(image2, image2_x1y1, image2_x2y2, (255, 0, 0), 8)
plt.imshow(image2)
```

```python
encoding1 = fc.face_encodings(image1)[0]
# print(encoding1)
encoding2 = fc.face_encodings(image2)[0]
# print(encoding2)
```

```python
result = fc.compare_faces([encoding1], encoding2)
```

```python
print(result)
```

```python
print(type(result))
```

```python
if(result[0] == False):
    result = 'Person is Different'
else:
    result = 'Person is Same'
```

```python
print(result)
```

```python
cv2.putText(image2, result, (30, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (156, 67, 81), 2)
plt.imshow(image2)
```
